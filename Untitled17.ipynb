```python
import pandas as pd

df = pd.read_csv("nepal_education_survey.csv")
print(df.columns)
```

```python
ct1 = pd.crosstab(df['gender'], df['school_type'])
print("ðŸ“Œ Crosstab: Gender vs School Type\n")
print(ct1)
```

```python
ct2 = pd.crosstab(df['province'], df['school_type'])
print("\nðŸ“Œ Crosstab: Province vs School Type\n")
print(ct2)
```

```python
ct3 = pd.crosstab(df['area'], df['internet_access'])
print("\nðŸ“Œ Crosstab: Area vs Internet Access\n")
print(ct3)
```

```python
ct4 = pd.crosstab(df['transport_mode'], df['school_type'])
print("\nðŸ“Œ Crosstab: Transport Mode vs School Type\n")
print(ct4)
```

```python
ct_percent = pd.crosstab(
    df['gender'],
    df['school_type'],
    normalize='index'
) * 100

print("\nðŸ“Œ Row Percentage Crosstab (Gender vs School Type)\n")
print(ct_percent.round(2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.heatmap(ct1, annot=True, fmt="d", cmap="Blues")
plt.title("Gender vs School Type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='gender', y='math_score')
plt.title("Math Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.violinplot(data=df, x='school_type', y='education_years')
plt.title("Years of Education by School Type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='internet_access', y='science_score')
plt.title("Science Score vs Internet Access")
plt.show()
```

```python
from scipy.stats import ttest_ind

group1 = df[df['gender'] == 'Male']['math_score']
group2 = df[df['gender'] == 'Female']['math_score']

t_stat, p_value = ttest_ind(group1, group2, nan_policy='omit')
print("T-test results:")
print("t-statistic =", t_stat)
print("p-value =", p_value)
```

```python
from scipy.stats import f_oneway

groups = [g['science_score'] for _, g in df.groupby('province')]

f_stat, p_value = f_oneway(*groups)
print("ANOVA results:")
print("F-statistic =", f_stat)
print("p-value =", p_value)
```

```python
import numpy as np

group_means = df.groupby('province')['math_score'].mean()
overall_mean = df['math_score'].mean()

ss_between = sum(df.groupby('province').size() * (group_means - overall_mean)**2)
ss_total = sum((df['math_score'] - overall_mean)**2)

eta_squared = ss_between / ss_total
print("Eta-squared (effect size):", eta_squared)
```

```python
numeric_cols = ['family_income_npr', 'education_years', 'math_score', 'science_score']

for col in numeric_cols:
    plt.figure(figsize=(12,4))

    # Histogram
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")

    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")

    plt.show()
```

```python
categorical_cols = ['province', 'area', 'gender', 'school_type',
                    'internet_access', 'transport_mode']

for col in categorical_cols:
    plt.figure(figsize=(7,4))
    sns.countplot(data=df, x=col)
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=30)
    plt.show()
```

```python
pairs = [
    ('gender', 'math_score'),
    ('area', 'science_score'),
    ('school_type', 'education_years')
]

for cat, num in pairs:
    plt.figure(figsize=(7,5))
    sns.boxplot(data=df, x=cat, y=num)
    plt.title(f"{num} by {cat}")
    plt.show()
```

```python
male = df[df['gender']=='Male']['math_score']
female = df[df['gender']=='Female']['math_score']

t_stat, p_val = ttest_ind(male, female, nan_policy='omit')
print("T-test (Math Score vs Gender):")
print("t-stat:", t_stat)
print("p-value:", p_val)
```
